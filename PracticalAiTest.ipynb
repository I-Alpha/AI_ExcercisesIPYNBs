```
#Q3a

def compute_expression(logic_expression, inter):                 # coverts string to logical epression and evaluates

    #Replace each variable with its value in dictionary. 
    logic_expression = ' '.join([inter.get(word,word) for word in logic_expression.split()])
 
    print('statement: ',logic_expression)
     
    #Replace the operators if they are caps
    logic_expression = logic_expression.replace("AND", "and")
    logic_expression = logic_expression.replace("OR", "or" )  
    logic_expression = logic_expression.replace("NOT", "not" ) 
    return eval(logic_expression)
    
Expression = "( A AND B ) or  B "
#interpr is a list of possible permutation of a given expression. 

interpr = {'A':['True','False'], 'B': ['False','True']}
interlist = [dict(zip(interpr, v)) for v in product(*interpr.values())]


for x in range(len(interlist)):
    print('Result :', compute_expression(Expression, interlist[x]),'\n')

#interpret the expression with eval(exp) and output its value
#print(Truths(['a', 'b'], ['not a','not b','(a and b)', 'a or b'], ints=False))
```

```
#3C

def  equivalence_test(logic_exp,inter_l):  #Ex4 function, uses Ex3 function to check statement using list of dicts 
    
    inter_len = len(inter_l)
    
    for x in range(inter_len):
        print('interpretation ',  x+1);
        if not((compute_expression(logic_exp[0], inter_l[x])) == (compute_expression(logic_exp[1], inter_l[x]))):
            print('not equal, therefore the logical expresssion pairs are not equivalent')
            return
        else:
            print('equal interpretations\n')
            
    print('these two logical expressions pairs are equivalent.')
    
    

#Ex4 Driver code 
#mAKE a list of dictionaries for holding possible permutations of truth variables. 
logicexp1 = 'A and B'
logicexp2 = 'not ( A ) and not ( B )'
interlist = {'A':['True','False'], 'B': ['False','True']}
interlist = [dict(zip(interlist, v)) for v in product(*interlist.values())]
interlist
expressns = [logicexp1,logicexp2]

equivalence_test(expressns,interlist)
    
```
