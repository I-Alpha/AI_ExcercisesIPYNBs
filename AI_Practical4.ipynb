```
%matplotlib inline
from matplotlib import pyplot as plt
import networkx as nx
```

```
#weighted graph 
G = nx.Graph()
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F'])
G.add_weighted_edges_from([
 ('A', 'B',7),
 ('A', 'F',14),
 ('A', 'C',9),
 ('B', 'C',10),
 ('B', 'D',15),
 ('C', 'F',2),
 ('C', 'D',11),
 ('F', 'E',9),
 ('D', 'E',6)
], weight='dist')
layout = nx.nx_pydot.graphviz_layout(G, prog='fdp')
nx.draw(G, pos=layout,with_labels=True)
edge_labels = nx.get_edge_attributes(G, 'dist')
nx.draw_networkx_edge_labels(G, pos=layout, edge_labels=edge_labels)
```

```
#exe One
paths =  [nx.shortest_path_length (G, source='A', weight = 'dist')]

print(paths)
```

```
#exe Two
paths_2 =  [nx.shortest_path(G, source='A', weight = 'dist')]
print(paths_2)
```
