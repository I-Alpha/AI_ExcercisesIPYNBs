```
%matplotlib inline
from matplotlib import pyplot as plt
import networkx as nx
```

```
#Exe 3
T = nx.DiGraph()
T.add_nodes_from(['S', 'A', 'B', 'C', 'D', 'E', 'F', 'G1','G2','G3'])
T.add_weighted_edges_from([
    ('S','A',5),
    ('S','B',9),
    ('S','D',6),
    ('A','G1',9),
    ('A','B',3),
    ('B','A',2),
    ('B','C',1),
    ('C','S',6),
    ('C','G2',5),
    ('C','F',7),
    ('D','S',1),
    ('D','C',2),
    ('D','E',2),
    ('E','G3',7),
    ('F','D',2),
    ('F','G3',8)
], weight ='new_dist')

layout = nx.nx_pydot.graphviz_layout(T, prog='neato')
nx.draw(T, pos=layout,with_labels=True)
edge_labels = nx.get_edge_attributes(T, 'new_dist')
nx.draw_networkx_edge_labels(T, pos=layout, edge_labels=edge_labels)
```

```
S_paths =  nx.shortest_path(T, source='S', target ='G1', weight = 'new_dist')
S_paths_len =  nx.shortest_path_length (T, source='S', target ='G1', weight = 'new_dist')
print(S_paths)
print(S_paths_len)
S_paths =  nx.shortest_path(T, source='S', target ='G2', weight = 'new_dist')
S_paths_len =  nx.shortest_path_length (T, source='S', target ='G2', weight = 'new_dist')
print(S_paths)
print(S_paths_len)
S_paths =  nx.shortest_path(T, source='S', target ='G3', weight = 'new_dist')
S_paths_len =  nx.shortest_path_length (T, source='S', target ='G3', weight = 'new_dist')
print(S_paths)
print(S_paths_len)
```
