```
#Ex2

def implies(a,b):   #input chars with ''. Returns truthtable with last argument a->b
   return Truths([a, b],['not ' + a,'not ' + b,'not '+ a + ' or '+ b], ints=False) 

    
def iff(a,b):   #input chars with ''. Returns truthtable with last argument a<->b
   return Truths([a, b],['not ' + a,'not ' + b, '(' + a + ' and ' + b + ') and (not ' + a + ' and not ' + b + ')'], ints=False)

print('Implication\n',implies('x','z'))
print('\nDouble Implication\n',iff('x','z'))
```

```
#Ex3

def compute_expression(logic_expression, inter):                 # coverts string to logical epression and evaluates

    #Replace each variable with its value in dictionary. 
    print("\n split expression",logic_expression.split())
    logic_expression = ' '.join([inter.get(word,word) for word in logic_expression.split()])
    print("\n",logic_expression)

    #Replace the operators if they are caps
    logic_expression = logic_expression.replace("AND", "and")
    logic_expression = logic_expression.replace("OR", "or" )  
    logic_expression = logic_expression.replace("NOT", "not" ) 
    
    #interpret the expression with eval(exp) and output its value
    return eval(logic_expression)


#Ex3 Driver code
# create a dict to hold values for replacing in Expression string
interpretation = {'A':'True',     
                  'B':'False'}

Expression = "A AND B AND ( NOT A OR NOT B )" # Characters must be separated by space
#                                                and match dict items in interpretation.
print(" Ex3......")
print('\n Result :',compute_expression(Expression, interpretation))

```

```

def eval_list(logic_expression,inter_l):  #Ex4 function, uses Ex3 function to check statement using list of dicts 
    
    inter_len = len(inter_l)
    f = 0    # initialise counters. 
    t = 0
    
    for x in range(inter_len):
        print ('\n item No.', x)
        if ((compute_expression(logic_expression, inter_l[x])) == True):
           t = t + 1
        else:
           f = f + 1
    print("\n\n|-------------------------------------------------------\n")
    if (f == 0):       
        print("\n This expression is a tautology.")
        return 1
    if (t == 0):
        print("\n This expression is a contradiction and is unsatisfiable")
        return -1
    else: 
        print("\n This expression is satisfiable.")
        return 0
    

#Ex4 Driver code 
#mAKE a list of dictionaries for holding possible permutations of truth variables.
interpretation_list = [{'A':'True','B':'False'},{'A':'False','B':'True'},{'A':'False','B':'False'},{'A':'True','B':'True'}]  
print("\n\n Ex4......")
print('\n Value returned :', eval_list(Expression,interpretation_list))

```
