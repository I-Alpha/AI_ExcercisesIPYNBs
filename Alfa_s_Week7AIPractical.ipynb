```
def add_node(g,u,a):
    #Add a node u labeled a in graph g
    g[u] = (a, [])
    
def add_edge(g, u, e, v):
    #Add an edge labeled e from u to v in graph g
    if (e,v) not in g[u][1]:
        g[u][1].append((e,v))
        
        
#Excercise 5  

word_list = nltk.word_tokenize("The child plays the fool")     #turning sentence into string list of individual words
cat_list = ["Det", "N", "V", "Det", "N"]                       # You can also just make a list of words. In this case it's 
                                                               #a list of categories. the element positions have to correspond
                                                               #to elements positions in word_list[] for iteration below to work.  
        
k = {}                                                         #initilizing dict object 
#k = dict()                                                    #or you can use this 


for i in range(1,len(word_list)+1):                                         #loop to iterate through word_list, similar to Excercise 4
    add_node(k, 'W%s' % i, ({'phon' : word_list[i-1], 'cat' : cat_list[i-1]}))  #adding nodes, words and categories from i=1


#word_list and cat_list to nodes in graph k  

#add the four edges 
add_edge(k, 'W2', 'det', 'W1')
add_edge(k, 'W3', 'suj', 'W2')     
add_edge(k, 'W3','obj', 'W5')
add_edge(k, 'W5', 'det', 'W4')

print("\nExercise 4 output...\n\niterated print of graph k:\n")  

for x in k:                                  #iterate through dict keys and print features in actual order
    print("Data in key",x,"in dict: " ,k[x],)

print("\n\nPrinting by writing just k makes collab print features in nodes alphabetically.")
print("Below you can see that cat feature now comes before phon which is not how its stored.\n\nk\n")

k
```

```
def add_node(g,u,a):
    #Add a node u labeled a in graph g
    g[u] = (a, [])
    
def add_edge(g, u, e, v):
    #Add an edge labeled e from u to v in graph g
    if (e,v) not in g[u][1]:
        g[u][1].append((e,v))
                      
def printGraphStuff(word_list,tag_list,feat_list,my_graph):          #function to print graph word_list and tag list and feature list loops
                                                            #because i cba to keep copy/paste/writing for-loops
    print('\nSentence:', " ".join(word_list),"\n")      
               
    for x in range(len(word_list)):                                        #iterate through dictionaries and print 
        print("Data in element",x,"in word_list: " ,word_list[x])  
    
    print("\n")

    for x in range(len(tag_list)):       
        print("Data in element",x,"in tag_list: " ,tag_list[x],)

    print("\n")
    
    for x in range(len(feat_list)):       
        print("Data in element",x,"in tag_list: " ,feat_list[x],)
    
    print('\nGraph:\n')
    for x in my_graph:                                  #iterate through dict keys and print features in actual order
        print("Data in key",x,"in dict: " ,my_graph[x],)

#Excercise 6


word_list = nltk.word_tokenize("The child plays the fool")      # Already explained what  nltk.word_tokenize does in excercise 5
tag_list = nltk.pos_tag(word_list)               # nltk.post_tag  gets the words and their category type from word_list. 
                                                 #so 'fool' would be NN for Noun. plays is a  verb so VBZ. 
feat_list  =[{'phon':n[0],'cat':n[1]} for n in tag_list]    #in every item in n of tag_list, get the element[0] and [1] and
                                                            #put them in features phon: and cat: respectively .
t_graph = {'W%s' % i:(feat_list[i],[])                      
           for i in range(len(tag_list))}

for i in range(1,len(tag_list)-1): #the last for-loop adds edges to the graph, similar to 
                                   #how I added edges in excercise 4, by iterating from 1 to (length of tag_list-1).                                
     add_edge(t_graph, 'W%s' % i,'SUC' , 'W%s' % (i+1))  #At each iteration, a 'SUC' edge is created from
                                                                                # the current node(i) to the the next one(i+1). 
    
                                                        

Altword_list = nltk.word_tokenize("She takes a glass")      #do the same thing as three lines above but for the alternate sentence. this part could be made into a function. 
Alttag_list = nltk.pos_tag(Altword_list)                          
Altfeat_list  =[{'phon':n[0],'cat':n[1]} for n in Alttag_list]                                                                   
Altt_graph = {'W%s' % i:(Altfeat_list[i],[])            
               for i in range(len(Alttag_list))}
         
for i in range(1,len(Alttag_list)-1):                                 
     add_edge(Altt_graph, 'W%s' % i,'SUC' , 'W%s' % (i+1)) 
    

    
print('t_graph-')
printGraphStuff(word_list,tag_list,feat_list,t_graph)    #call my function to print components used to build graph.
print('\n\nAltt_graph-\n')                                         
printGraphStuff(Altword_list,Alttag_list,Altfeat_list,Altt_graph)    

# In summary, the code does what i did in excercise 5, but using  .post_tag to get list of categories and, at the end, add edges using items from tag_list.
```
